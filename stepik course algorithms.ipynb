```python
def fib(n):
    su=0
    if n<=1:
        su = n
        return (su)
    else :
        su = fib(n-1) + fib(n-2)
        return (su)
```

```python
fib(35)
# уже секунд 5
```

```python
def fibar(n):
    flist=[0,1]
    for i in range(1,n+1):
        fval=(flist[i-1]+flist[i-2])
        flist.append(fval)
        flist[i]=(flist[i-1]+flist[i-2])
    return flist[n]
```

```python
fibar(2)
#считает мнгновенно
```

```python
list(map(fibar, test_l))
```

```python
def naiveGCD(a,b):
    gcd=1
    for d in range(2,min(a,b)+1):
        if a%d==0 and b%d==0:
            gcd=d
    return (gcd)
```
